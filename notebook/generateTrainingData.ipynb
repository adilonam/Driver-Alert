```python
import numpy as np
from scipy import signal
RATE = 22050
sos = signal.butter(5, [50, 5000], 'bandpass', fs=RATE, output='sos') # can filter here or when you train


def get_audio(audio):
    # converts data into 3 second chunk 
   
    try:
        
        audio = audio[0:RATE*3]
        
        #audio = signal.sosfilt(sos, audio) 

    except Exception as e:
            print("Error encountered for a file")
            return None

     
    return audio
```

```python
## Generate 3-second bytes in the form of numpy arrays of the car and siren sounds saved in text files (generated from convertData2Txt)

import pandas as pd
import numpy as np
import os
import librosa
from scipy import signal

sr = RATE
samples = sr*3

import time
start_time = time.time()



filename = r'C:\Users\Gabe\Documents\SeniorDesign\Data\Senior Design data training\CarBackground3sec.txt'

df = pd.DataFrame()
for chunk in pd.read_csv(filename, header=None, delim_whitespace=True, chunksize=1000):
    data = pd.concat([df, chunk], ignore_index=True) 
    
data = np.array(data)

carNoise = []
for index in range(0, data.shape[0]):
    carNewSR = signal.resample(data[index,:], samples)
    carNoise.append(carNewSR)
    
print(np.array(carNoise).shape)
    
print("--- %s seconds ---" % (time.time() - start_time))



filename = r'C:\Users\Gabe\Documents\SeniorDesign\Data\Senior Design data training\Siren3sec.txt'

df = pd.DataFrame()
for chunk in pd.read_csv(filename, header=None, delim_whitespace=True, chunksize=1000):
    data = pd.concat([df, chunk], ignore_index=True)    
data = np.array(data)


sirenNoise = []
for index in range(0, data.shape[0]):
    sirenNewSR = signal.resample(data[index,:], samples)
    sirenNoise.append(sirenNewSR)
    
print(np.array(sirenNoise).shape)

    
print("--- %s seconds ---" % (time.time() - start_time))
    
```

```python
# generates numpy array of environment sounds from UrbanSounds8k dataset

# Load various imports 
import pandas as pd
import os
import librosa

# Set the path to the full UrbanSound dataset 
fulldatasetpath = 'C:/Users/Gabe/Documents/PYTHON/DataSets/UrbanSound8K/audio/'

metadata = pd.read_csv('C:/Users/Gabe/Documents/PYTHON/Udacity-ML-Capstone-master/UrbanSound Dataset sample/metadata/UrbanSound8K_SirenVsNot.csv')

sirens = []
background = []

# Iterate through each sound file and extract the features 
for index, row in metadata.iterrows():
    
    file_name = os.path.join(os.path.abspath(fulldatasetpath),'fold'+str(row["fold"])+'/',str(row["slice_file_name"]))
    
    class_label = row["class_name"]
    
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    if len(audio)<(sample_rate*3):
        print("excluding ", file_name)
        continue
        
    data = get_audio(audio)
    
    if class_label == 'siren':
        sirens.append(data)
    else:
        background.append(data)
    
```

```python
## Save data as numpy structures

# combines UrbanSounds8k sirens with text file sirens
siren = np.array(sirens)
siren2 = np.array(sirenNoise)
sirenNoise = np.concatenate((siren2, siren))

backNoise = np.array(background)

carNoise = np.array(carNoise)

np.save(sirenFile, sirenNoise)
np.save(backFile, backNoise)
np.save(carFile, carNoise)
```
